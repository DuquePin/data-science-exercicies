````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import statsmodels.formula.api as smf
````

````python
df = pd.read_csv('./previsao_de_renda_II.csv', index_col=0)
df.head()
````

## Avalie valores missing e substitua pela mdia

Neste exerccio no vamos nos preocupar com valores *missing*. Substitua-os pela mdia.

````python
df.info()
````

````python
df['tempo_emprego'] = df['tempo_emprego'].fillna(df['tempo_emprego'].mean())

df.info()
````

### Base de testes

Separe os trs ltimos meses como base *holdout* (base de teste).

````python
df['data'] = pd.to_datetime(df.index)
df['data']
````

````python
df['mes'] = [row.month for row in df['data']]
df['mes']
````

````python
df_teste = df[(df['mes']==10) | (df['mes']==11) | (df['mes']==12)]
df_teste['mes'].value_counts()
````

````python
df_treino = df[(df['mes']!=10) & (df['mes']!=11) & (df['mes']!=12)]
df_treino['mes'].value_counts()
````

### Primeiro modelo
 No use *index* e no use a data. Avalie o modelo com o $R^2-ajustado$

````python
reg1 = smf.ols('renda ~ sexo + posse_de_imovel + estado_civil + idade + tempo_emprego', data=df_treino).fit()

res1 = reg1.summary()

res1.tables[0]
````

### Anlise de perfil

Categorize a varivel ```tempo_emprego``` em 20 quantis (cada categoria deve ter aproximadamente a mesma quantidade de observaes) e faa uma tabela de perfil (armazene como dataframe) para esta categorizao:

- Cada linha  uma categoria
- Uma coluna para a quantidade de observaes de cada categoria
- Uma coluna para a mdia do log-renda para cada categoria
- Uma coluna para a mdia do tempo de emprego para cada categoria

Faa um grfico da mdia do tempo de emprego pela mdia do log-renda. Insira uma clula de texto e comente se te parece que alguma das transformaes vistas no mdulo deve fornecer bons resultados.

````python
perfil = df_treino[['tempo_emprego', 'renda']].copy()

perfil['cat_tempo_emprego'] = pd.qcut(x=df_treino['tempo_emprego'], q=20, duplicates='drop')

perfil.head(5)
````

````python
min_count = perfil['cat_tempo_emprego'].value_counts().min()

perfil_balanced = (perfil
                   .groupby('cat_tempo_emprego', observed=False)[['tempo_emprego', 'renda', 'cat_tempo_emprego']]
                   .apply(lambda x: x.sample(n=min_count, random_state=42))
                   .reset_index(drop=True)
                  )

gp = (perfil_balanced
      .groupby(by='cat_tempo_emprego', observed=False)[['tempo_emprego', 'renda']]
      .agg(tempo_emprego_medio=('tempo_emprego', 'mean'), renda_media=('renda', 'mean'), numeros_observacoes=('tempo_emprego', 'count'))
     )

gp['log_renda_media'] = np.log(gp['renda_media'])

gp
````

````python
perfil['log_renda_media'] = perfil['cat_tempo_emprego'].map(gp['log_renda_media'])
perfil['tempo_emprego_medio'] = perfil['cat_tempo_emprego'].map(gp['tempo_emprego_medio'])
perfil['renda_media'] = perfil['cat_tempo_emprego'].map(gp['renda_media'])
perfil['log_renda'] = np.log(perfil['renda'])
perfil['log_tempo_emprego'] = np.log(perfil['tempo_emprego'])

perfil.head(5)
````

````python
sns.scatterplot(data=perfil, x='tempo_emprego_medio', y='log_renda_media')
plt.tight_layout()
plt.show()
````

````python
sns.scatterplot(data=perfil, x='tempo_emprego', y='log_renda')
plt.tight_layout()
plt.show()
````
