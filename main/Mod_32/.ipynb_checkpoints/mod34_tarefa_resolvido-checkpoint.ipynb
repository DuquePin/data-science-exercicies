# Regressão III - Tarefa

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import statsmodels.formula.api as smf
````

````python
df = pd.read_csv('./previsao_de_renda_II.csv', index_col=0)
df.head()
````

## Avalie valores missing e substitua pela média

Neste exercício não vamos nos preocupar com valores *missing*. Substitua-os pela média.

````python
df['tempo_emprego'] = df['tempo_emprego'].fillna(df['tempo_emprego'].mean())

df.info()
````

### Base de testes

Separe os três últimos meses como base *holdout* (base de teste).

````python
df['data'] = pd.to_datetime(df.index)

df['mes'] = [row.month for row in df['data']]

holdout = df[(df['mes']==10) | (df['mes']==11) | (df['mes']==12)]

df = df[(df['mes']!=10) & (df['mes']!=11) & (df['mes']!=12)]

holdout.head(5)
````

### Primeiro modelo
 Não use *index* e não use a data. Avalie o modelo com o $R^2-ajustado$

````python
reg1 = smf.ols('renda ~ sexo + posse_de_imovel + estado_civil + idade + tempo_emprego', data=df).fit()

print(f'R² ajustado do modelo igual a {reg1.rsquared_adj:.2%}')
````

### Análise de perfil

Categorize a variável ```tempo_emprego``` em 20 quantis (cada categoria deve ter aproximadamente a mesma quantidade de observações) e faça uma tabela de perfil (armazene como dataframe) para esta categorização:

- Cada linha é uma categoria
- Uma coluna para a quantidade de observações de cada categoria
- Uma coluna para a média do log-renda para cada categoria
- Uma coluna para a média do tempo de emprego para cada categoria

Faça um gráfico da média do tempo de emprego pela média do log-renda. Insira uma célula de texto e comente se te parece que alguma das transformações vistas no módulo deve fornecer bons resultados.

````python
perfil = df[['tempo_emprego', 'renda']].copy()

perfil['cat_tempo_emprego'] = pd.qcut(x=df['tempo_emprego'], q=20, duplicates='drop')

perfil.head(5)
````

````python
min_count = perfil['cat_tempo_emprego'].value_counts().min()

perfil_balanced = (perfil
                   .groupby('cat_tempo_emprego', observed=False)[['tempo_emprego', 'renda', 'cat_tempo_emprego']]
                   .apply(lambda x: x.sample(n=min_count, random_state=42))
                   .reset_index(drop=True)
                  )

gp = (perfil_balanced
      .groupby(by='cat_tempo_emprego', observed=False)[['tempo_emprego', 'renda']]
      .agg(tempo_emprego_medio=('tempo_emprego', 'mean'), renda_media=('renda', 'mean'), numeros_observacoes=('tempo_emprego', 'count'))
     )

gp['log_renda_media'] = np.log(gp['renda_media'])

gp
````

````python
gp.plot(x='tempo_emprego_medio', y='log_renda_media', kind='line')
plt.show()
````

````python
perfil['log_renda_media'] = perfil['cat_tempo_emprego'].map(gp['log_renda_media'])
perfil['tempo_emprego_medio'] = (perfil['cat_tempo_emprego'].map(gp['tempo_emprego_medio'])).astype(np.float64)
perfil['renda_media'] = perfil['cat_tempo_emprego'].map(gp['renda_media'])
perfil['log_renda'] = np.log(perfil['renda'])
perfil['log_tempo_emprego'] = np.log(perfil['tempo_emprego'])

perfil.head(5)
````

````python
sns.scatterplot(data=perfil, x='tempo_emprego_medio', y='log_renda_media')
plt.tight_layout()
plt.show()
````

Conforme observado nos gráficos acima, a relação entre **log_renda_media** e **tempo_emprego_medio** parece ser linear, o que significa que uma transformação nos dados pode não alterar a relação que essas variáveis já possuem.

### Linearização

Experimente pelo menos duas técnicas de linearização vistas ao longo do módulo para melhorar o ajuste da variável ```tempo_emprego```. Compare o $R^2-ajustado$ e avalie se alguma delas apresentou resultados relevantes com relação ao primeiro modelo.


````python
# Transformação em log

X = np.log(perfil['tempo_emprego_medio'])
y = perfil['log_renda_media']

sns.scatterplot(x=X, y=y)
plt.tight_layout()
plt.show()
````

````python
reg2 = smf.ols('np.log(renda) ~ sexo + posse_de_imovel + estado_civil + idade + np.log(tempo_emprego)', data=df).fit()

print(f'R-quadrado ajustado: {reg2.rsquared:.2%}')
````

````python
X = np.power(perfil['tempo_emprego_medio'], 2)
y = perfil['log_renda_media']

sns.scatterplot(x=X, y=y)
plt.tight_layout()
plt.show()
````

````python
# Transformação em exponencial

reg3 = smf.ols('np.log(renda) ~ sexo + posse_de_imovel + estado_civil + idade + I(tempo_emprego**2)', data=df).fit()

print(f'R-quadrado: {reg3.rsquared:.2%}')
````

````python
print('Resultados')
print(f'R² Ajustado 1º Modelo: {reg1.rsquared_adj:.2%}')
print(f'R² Ajustado 2º Modelo: {reg2.rsquared_adj:.2%}')
print(f'R² Ajustado 3º Modelo: {reg3.rsquared_adj:.2%}')
````

Como podemos observar acima, aplicar as transformações logarítimicas e exponenciais na variável **tempo_emprego** melhorou o desempenho do modelo.

### Avaliando o modelo

Avalie os resultados dos modelos construidos na base de testes.

````python
for i, modelo in enumerate([reg1, reg2, reg3]):
    # Prever os valores na base de testes
    y_pred = modelo.get_prediction(holdout).predicted_mean
    
    # Calcular o R² na base de testes
    SSE = np.sum((holdout['renda'] - y_pred) ** 2)
    SST = np.sum((holdout['renda'] - np.mean(holdout['renda'])) ** 2)
    R2 = 1 - (SSE / SST)
    
    # Calcular o R² ajustado
    n = len(holdout['renda'])  
    p = len(modelo.params) - 1
    R2_ajustado = 1 - ( (1 - R2) * (n - 1) / (n - p - 1) )
    
    print(f"R² ajustado na base de testes do {i+1}º Modelo: {R2_ajustado:.2%}")
````
