# GradientBoosting Machine

O algoritmo **Gradient Boosting Machine** é uma outra técnica de **ensemble**, na qual uma floresta de árvores de decisão é gerada para criar uma resposta final. Pode-se considerar esse algoritmo como uma especialização do algorimto de **Random Forest**. Dado um conjunto de dados $D$ com $n$ linhas (observações) e $m$ colunas (features), calcula-se a média dos valores da coluna target ($y$), esse valor é usado na primeira iteração de treinamento das $T$ árvores de decisões. A partir da média, denominados $\overline y$, calcula-se o resíduo $\epsilon_1$ dado por: $\epsilon_1 = y - \overline y$

Após o cálculo dos resíduos, as árvores de decisão são treinadas até uma profundidade máxima $P$, através de amostragens com reposição (**bootstrap sample**) e elas são usadas para prever valores $\hat y_1$ usando os resíduos $\epsilon_1$. Após isso, para garantir a robustez do modelo e evitar **overfitting**, um fator arbitrário é aplicado para calcular os resíduos dos modelos. Esse fator é denominado **learning_rate**, representando por $\eta$ e é responsável por controlar o quão bem os modelos se adequam aos dados. Os resíduos dessa etapa são calculados pela expressão: $\epsilon_2 = \overline y + (\eta \cdot \hat y_1)$

O processo é repetido para próxima iteração, com as funções responsáveis por calcular os resíduos (**loss functions**) sendo incrementadas a cada novo valor de preditores $\hat y_i$. A função genérica é dada por:

$$\epsilon_i = \overline y + \sum^p_{i=1} (\hat y_i \cdot \eta)$$

Ao adotar essa estratégia, o algoritmo **Gradient Boosting Machine** é capaz de, a partir da média dos valores target, "caminhar" até os valores mais ajustados, evitando **overfitting** no modelo e criando uma resposta final mais acurada e robusta.

# Diferenças com Adapative Boosting Machine

As principais diferenças dos algoritmos são:

**Floresta**

O algoritmo AdaBoost usa uma floresta de stumps (árvores de decisão de profundidade 1) para seus **weak learners**, enquanto o algoritmo GradientBoost usa uma floresta de árvores de decisão genéricas, mas comumente limitadas a uma profundidade $P$.

**Treinamento**

A primeira etapa de treinamento dos modelos, no caso do AdaBoost, é realizada através das previsões usando os **pesos** pré-definidos de cada linha, enquanto o GradientBoost utiliza a média dos valores target.

**Ajuste**

Ao avançar pelas iterações, o modelo AdaBoost ajusta os **pesos** com base no desempenho do melhor modelo da iteração anterior, focando nos erros de previsão dos modelos, enquanto o modelo GradientBoost ajusta a **função de perda** com o uso do parâmetro de **learning_rate** para caminhar a partir da média até valores mais precisos.

**Sensibilidade a Outliers**

AdaBoost tende a ser mais sensível a **outliers**, pois as instâncias mal classificadas (incluindo outliers) recebem pesos maiores a cada iteração, fazendo com que o modelo foque mais nelas. Isso pode levar a um **overfitting** nos outliers. Em contrapartida, GradientBoosting é menos sensível aos outliers, já que o ajuste é feito de forma gradual e controlada pelo **learning_rate**, evitando que os outliers dominem o treinamento.

**Interpretação de Erros**

Outra diferença está na maneira como os erros são tratados. O AdaBoost ajusta diretamente os pesos das instâncias mal classificadas, enfatizando os erros mais difíceis de corrigir. Por outro lado, o GradientBoosting ajusta os **resíduos** (a diferença entre as previsões atuais e os valores reais) a cada iteração, minimizando a função de perda de forma global, resultando em uma abordagem mais equilibrada para melhorar a precisão do modelo.

**Dados Faltantes**

Uma grande diferença entre os algoritmos é a forma como lidam com dados **missing**, o algoritmo de GradientBoosting tem uma forma nativa para lidar com linhas com dados faltante, no entanto, o algoritmo AdaBoost não possui tal ferramenta, sendo necessário um pré-processamento de dados antes do treino do modelo.

# Implementação dos Algoritmos

```python
from timeit import timeit

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, GradientBoostingRegressor, AdaBoostRegressor

from sklearn.datasets import make_classification, make_regression

from sklearn.model_selection import cross_val_score
```

## Caso de Classificação

```python
%%time
# Criando base de dados 
X, y = make_classification(n_samples=10000, random_state=412)

# Implementando GradientBoost
gbc_model = GradientBoostingClassifier(n_estimators=100, learning_rate=.1, max_depth=3, random_state=412)
scores_gbc = cross_val_score(gbc_model, X, y, cv=10)

# Capturando tempo de execução do algoritmo
exet_gbc = timeit(stmt='cross_val_score(gbc_model, X, y, cv=10)', globals=globals(),  number=1)

# Implementando AdaBoost
abc_model = AdaBoostClassifier(n_estimators=100, algorithm='SAMME', random_state=412)
scores_abc = cross_val_score(abc_model, X, y, cv=10)

# Capturando tempo de execução do algoritmo
exet_abc = timeit(stmt='cross_val_score(abc_model, X, y, cv=10)', globals=globals(),  number=1)

# Exbinido gráfico de linha comparativo
fig, ax = plt.subplots()

sns.lineplot(scores_gbc, ax=ax, label='GradientBoostingClassifier')
sns.lineplot(scores_abc, ax=ax, label='AdaBoostClassifier')

plt.xlabel('Folds')
plt.ylabel('Scores')
plt.title('Desempenho Algoritmos GBC vs. ABC')

plt.tight_layout()
plt.show()
```

```python
print(f'Score médio do algoritmo GradientBoostingClassifer: {scores_gbc.mean():.2f}\nScore médio do algoritmo AdaBoostClassifer: {scores_abc.mean():.2f}')
print(f'Tempo de execução algoritmo GradientBoostingClassifer: {exet_gbc:.2f} s\nTempo de execução algoritmo AdaBoostClassifier: {exet_abc:.2f} s')
```

# Insights

Como é possível observar acima, o algoritmo GBC (**GradientBoostingClassifier**) tem uma precisão média levemente acima do algoritmo ABC (**AdaBoostClassifier**), no entanto, o mesmo executou consideravelmente mais rápido do que o algoritmo GBC, uma diferença de aproximadamente 33 segundos. O motivo disso se deve porque o algoritmo GBC tem um complexidade maior (semelhante a complexidade de um algoritmo **Random Forest**) e por isso, consome mais recursos computacionais.

Vale salientar, ambos algoritmos são execelentes ferramentas e para determinar seus usos em casos especificos, é necessário um estudo do caso e analisar quais são as necessidades do mesmo. Por exemplo, se o algoritmo irá lidar com uma base de dados muito grande, pode ser vantajoso utilizar o algoritmo ABC, no entanto, se a qualidade dos dados não pode ser assegurada (existe um risco muito grande de dados **missing** ou **outliers**) é recomendável utilizar o algoritmo GBC, pois esse não é prejudicado por outilers da forma como o algoritmo ABC, além de possuir uma forma nativa de lidar com dados **missing**, com o custo adicional de consumir mais recursos computacionais que o algoritmo ABC.

## Caso de Regressão

```python
%%time
# Criando base de dados 
X, y = make_regression(n_samples=1000, random_state=412)

# Implementando GradientBoost
gbr_model = GradientBoostingRegressor(n_estimators=100, learning_rate=.1, max_depth=3, random_state=412)
scores_gbr = cross_val_score(gbr_model, X, y, cv=10)

# Capturando tempo de execução do algoritmo
exet_gbr = timeit(stmt='cross_val_score(gbr_model, X, y, cv=10)', globals=globals(),  number=1)

# Implementando AdaBoost
abr_model = AdaBoostRegressor(n_estimators=100, loss='linear', random_state=412)
scores_abr = cross_val_score(abr_model, X, y, cv=10)

# Capturando tempo de execução do algoritmo
exet_abr = timeit(stmt='cross_val_score(abr_model, X, y, cv=10)', globals=globals(),  number=1)

# Exbinido gráfico de linha comparativo
fig, ax = plt.subplots()

sns.lineplot(scores_gbr, ax=ax, label='GradientBoostingRegressor')
sns.lineplot(scores_abr, ax=ax, label='AdaBoostRegressor')

plt.xlabel('Folds')
plt.ylabel('Scores')
plt.title('Desempenho Algoritmos GBR vs. ABR')

plt.tight_layout()
plt.show()

print(f'Score médio do algoritmo GradientBoostingRegressor: {scores_gbr.mean():.2f}\nScore médio do algoritmo AdaBoostRegressor: {scores_abr.mean():.2f}')
print(f'Tempo de execução algoritmo GradientBoostingRegressor: {exet_gbr:.2f} s\nTempo de execução algoritmo AdaBoostRegressor: {exet_abr:.2f} s')
```

# Insights

Como é possível observar acima, o mesmo padrão do caso de classificação se repete, o algoritmo ABR realiza suas previsões mais rapidamente quando comparado ao algoritmo GBR, no entanto, o algoritmo GBR é mais preciso.

# Hiper-parâmetros

Os cinco hiper-parâmetros mais importantes do algoritmo **GradientBoostingClassifier** são:

**n_estimators**

Número de árvores a serem construídas. Um valor maior pode melhorar o desempenho, mas também aumenta o risco de sobreajuste e o tempo de treinamento.

**learning_rate**

Fator de redução aplicado às contribuições de cada árvore. Valores menores tornam o treinamento mais robusto, mas exigem mais árvores para obter o mesmo desempenho.

**max_depth**

Profundidade máxima das árvores individuais. Limitar a profundidade ajuda a evitar sobreajuste, mantendo o modelo mais simples.

**subsample**

Proporção dos dados de treinamento a serem usados em cada iteração. Valores menores que 1.0 podem reduzir o sobreajuste, introduzindo aleatoriedade no treinamento.

**tol**

A tolerância do modelo, quando a função de perda não estiver melhorando o modelo por esse valor, o algoritmo finaliza a executação mais cedo, permitindo um resultado mais rápido.



## Stochastic Gradient Boosting Machine

O **Stochastic Gradient Boosting Machine (SGBM)** é uma variação do algoritmo tradicional de **Gradient Boosting**. Ao ajustar o hiperparâmetro **subsample**, é possível aplicar a estratégia estocástica ao modelo GBM. O termo "Stochastic" significa **estocástico**, ou seja, algo que envolve aleatoriedade.

A principal diferença entre o SGBM e o GBM tradicional está no fato de que, no SGBM, o modelo utiliza uma **amostragem aleatória sem reposição** a cada iteração. Isso significa que, em vez de usar todo o conjunto de dados, o modelo seleciona aleatoriamente uma fração das observações, definida pelo parâmetro subsample, para treinar cada árvore. Essa abordagem reduz a probabilidade de **overfitting**, tornando o modelo mais robusto, já que a aleatoriedade introduzida pela amostragem diversifica os dados usados para o treinamento, sem reutilizar as mesmas observações em cada iteração.

Essa técnica melhora a generalização do modelo, evitando que ele se ajuste excessivamente aos dados de treinamento e aumentando sua capacidade de performar bem em novos dados. Além de torná-lo mais rápido por utilizar base de dados menores.
