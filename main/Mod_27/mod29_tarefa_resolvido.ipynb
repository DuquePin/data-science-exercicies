## Número de grupos

Utilize as técnicas vistas em aula que te ajudem a decidir pelo número de grupos a ser utilizados.

Antes de prosseguirmos para a modelagem em si, vamos padronizar as variáveis de forma a facilitar o trabalho do modelo.

```python
wdf_pad = pd.DataFrame(StandardScaler().fit_transform(wdf.drop('R_groups', axis=1)), 
                       columns=wdf.drop('R_groups', axis=1).columns, 
                       index=wdf.index)
wdf_pad['R_groups'] = wdf['R_groups']
wdf_pad.head(5)
```

Para definir a quantidade de **clusters** a ser utilizado, vamos aplicar o método da silhueta.

```python
siluetas = []
max_clusters = 15

X = wdf_pad.drop('R_groups', axis=1)

for n_clusters in range(2, max_clusters + 1):
    km = KMeans(n_clusters=n_clusters, random_state=412).fit(X)
    siluetas.append(silhouette_score(X, km.labels_))
    nomes_grupos = ['grupo_' + str(g) for g in range(n_clusters)]
    wdf_pad[str(n_clusters) + '_grupos'] = pd.Categorical.from_codes(km.labels_, categories=nomes_grupos)

df_silueta = pd.DataFrame({'n_clusters': list(range(2, max_clusters + 1)), 'silueta_média': siluetas})
df_silueta.plot.line(x='n_clusters', y='silueta_média', marker='o')
plt.show()
```

Observando os gráficos acima, vamos selecionar 2, 3 e 4 grupos para testar diferentes abordagens na modelagem.

## Avaliação dos grupos

Faça uma análise descritiva para pelo menos duas soluções de agrupamentos (duas quantidades diferentes de grupos) sugeridas no item anterior, utilizando as variáveis que estão no escopo do agrupamento.
- Com base nesta análise e nas análises anteriores, decida pelo agrupamento final. 
- Se puder, sugira nomes para os grupos.

```python
# Função para avaliar um determinado número de grupos
def avaliar_grupo(grupo):
    ax = wdf_pad.groupby([grupo, 'R_groups'], observed=False)['R_groups'].count().unstack().plot.bar()
    ax.legend(loc='lower center', bbox_to_anchor=(0.5, -.5), ncol=3, fancybox=True, shadow=True)
    plt.title('Distribuição de Grupos ' + grupo)
    plt.show()
```

### Agrupando em 2 Grupos

O problema destacado anteriormente é evidente no gráfico abaixo: a classe dominante predomina em ambos os grupos, o que compromete a capacidade do modelo de diferenciar clientes com base em seu comportamento no site, tornando-o inviável para aplicações práticas. Apesar disso, observa-se que um dos grupos conseguiu captar a maioria dos novos visitantes que não concluíram a compra. No entanto, essa diferenciação é insuficiente para tornar o modelo aplicável de forma eficaz em um cenário real.

```python
avaliar_grupo('2_grupos')
```

### Agrupando em 3 Grupos

O mesmo problema persiste, embora, neste caso, haja uma distribuição ligeiramente mais equilibrada das demais classes em comparação com a distribuição anterior.

```python
avaliar_grupo('3_grupos')
```

### Agrupando em 4 Grupos

O mesmo padrão se repete, com dois grupos concentrando a maioria das classes, refletindo uma distribuição similar à observada nos agrupamentos anteriores de 2 e 3 grupos. Esse comportamento sugere que o modelo continua a agrupar as classes de maneira desequilibrada, concentrando-as em alguns grupos principais.

```python
avaliar_grupo('4_grupos')
```
