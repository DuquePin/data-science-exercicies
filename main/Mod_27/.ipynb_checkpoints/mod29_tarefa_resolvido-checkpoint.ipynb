## Nmero de grupos

Utilize as tcnicas vistas em aula que te ajudem a decidir pelo nmero de grupos a ser utilizados.

Antes de prosseguirmos para a modelagem em si, vamos padronizar as variveis de forma a facilitar o trabalho do modelo.

```python
wdf_pad = pd.DataFrame(StandardScaler().fit_transform(wdf.drop('R_groups', axis=1)), 
                       columns=wdf.drop('R_groups', axis=1).columns, 
                       index=wdf.index)
wdf_pad['R_groups'] = wdf['R_groups']
wdf_pad.head(5)
```

Para definir a quantidade de **clusters** a ser utilizado, vamos aplicar o mtodo da silhueta.

```python
siluetas = []
max_clusters = 15

X = wdf_pad.drop('R_groups', axis=1)

for n_clusters in range(2, max_clusters + 1):
    km = KMeans(n_clusters=n_clusters, random_state=412).fit(X)
    siluetas.append(silhouette_score(X, km.labels_))
    nomes_grupos = ['grupo_' + str(g) for g in range(n_clusters)]
    wdf_pad[str(n_clusters) + '_grupos'] = pd.Categorical.from_codes(km.labels_, categories=nomes_grupos)

df_silueta = pd.DataFrame({'n_clusters': list(range(2, max_clusters + 1)), 'silueta_mdia': siluetas})
df_silueta.plot.line(x='n_clusters', y='silueta_mdia', marker='o')
plt.show()
```

Observando os grficos acima, vamos selecionar 2, 3 e 4 grupos para testar diferentes abordagens na modelagem.

## Avaliao dos grupos

Faa uma anlise descritiva para pelo menos duas solues de agrupamentos (duas quantidades diferentes de grupos) sugeridas no item anterior, utilizando as variveis que esto no escopo do agrupamento.
- Com base nesta anlise e nas anlises anteriores, decida pelo agrupamento final. 
- Se puder, sugira nomes para os grupos.

```python
# Funo para avaliar um determinado nmero de grupos
def avaliar_grupo(grupo):
    ax = wdf_pad.groupby([grupo, 'R_groups'], observed=False)['R_groups'].count().unstack().plot.bar()
    ax.legend(loc='lower center', bbox_to_anchor=(0.5, -.5), ncol=3, fancybox=True, shadow=True)
    plt.title('Distribuio de Grupos ' + grupo)
    plt.show()
```

### Agrupando em 2 Grupos

O problema destacado anteriormente  evidente no grfico abaixo: a classe dominante predomina em ambos os grupos, o que compromete a capacidade do modelo de diferenciar clientes com base em seu comportamento no site, tornando-o invivel para aplicaes prticas. Apesar disso, observa-se que um dos grupos conseguiu captar a maioria dos novos visitantes que no concluram a compra. No entanto, essa diferenciao  insuficiente para tornar o modelo aplicvel de forma eficaz em um cenrio real.

```python
avaliar_grupo('2_grupos')
```

### Agrupando em 3 Grupos

O mesmo problema persiste, embora, neste caso, haja uma distribuio ligeiramente mais equilibrada das demais classes em comparao com a distribuio anterior.

```python
avaliar_grupo('3_grupos')
```

### Agrupando em 4 Grupos

O mesmo padro se repete, com dois grupos concentrando a maioria das classes, refletindo uma distribuio similar  observada nos agrupamentos anteriores de 2 e 3 grupos. Esse comportamento sugere que o modelo continua a agrupar as classes de maneira desequilibrada, concentrando-as em alguns grupos principais.

```python
avaliar_grupo('4_grupos')
```
