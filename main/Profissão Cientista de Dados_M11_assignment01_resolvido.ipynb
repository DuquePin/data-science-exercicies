### 3. Separe os dados em validação e teste

````python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)

x_train.head(5)
````

````python
x_test.head(5)
````

### 4. Treine duas árvores, uma com profundidade máxima = 8, outra com profundidade máxima = 2.

Como uma forma de melhor visualizar o impacto da profundidade na criação da árvore vamos considerar um intervalo de profundidades entre 2 e 32.

````python
# Criando os modelos
regtrees = [DecisionTreeRegressor(max_depth=mxd, random_state=100) for mxd in range(2, 33)]
regtrees
````

````python
# Treinando os modelos e criando previsões
predictations = []

for regtree in regtrees:
    regtree.fit(x_train, y_train)
    predictations.append(regtree.predict(x_test))   
````

````python
# Visualizando modelo de profundidade máxima 2
dot_regtree2 = export_graphviz(regtrees[0], out_file=None,  
                               feature_names=X.columns,   
                               filled=True, rounded=True,  
                               special_characters=True)
graph2 = graphviz.Source(dot_regtree2)
graph2
````

````python
# Visualizando modelo de profundidade 8
dot_regtree8 = export_graphviz(regtrees[7], out_file=None,  
                               feature_names=X.columns,   
                               filled=True, rounded=True,  
                               special_characters=True)
graph8 = graphviz.Source(dot_regtree8)
graph8
````

### 5. Calcule o MSE de cada uma das árvores do item anterior para a base de treinamento e para a base de testes.

Com o intuito de melhor analisar o desempenho dos modelos, vamos definir uma função para calcular o r2 do mesmo.

````python
# Função para calcular r2
def r2(y, y_pred):
    return 1 - (len(y) * mse(y, y_pred) / np.sum([pow(i - np.mean(y), 2) for i in y]))

# Criando tabela de comparação entre modelos
comp = pd.DataFrame({'max_depth': [value for value in range(2, 33)], 
                     'mean_square_error': [mse(y_test, predicts[index]) for index in range(31)],
                     'r^2': [r2(y_test, predicts[index]) for index in range(31)]})
comp.head(5)
````

````python
# Criando figura e eixo
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotando no eixo um gráfico de linha com o r^2
sns.lineplot(x='max_depth', y='r^2', data=comp, ax=ax1, color='b')
ax1.set_ylabel('r^2', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Criando um novo eixo y
ax2 = ax1.twinx()

# Plotando no segundo eixo y um gráfico de linha com o erro quadrático médio
sns.lineplot(x='max_depth', y='mean_square_error', data=comp, ax=ax2, color='r')
ax2.set_ylabel('mse', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Subdividindo eixo x em 32 partes
ax1.set_xlabel('max_depth')
ax1.set_xticks([value for value in range(33)])

# Visualizando figura
plt.show()
````
