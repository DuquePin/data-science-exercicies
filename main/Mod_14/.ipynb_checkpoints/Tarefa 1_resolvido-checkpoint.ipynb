- Gerar gráficos referentes a base de dados de março, abril, maio, junho e dezembro

- Como automatizar o processo? O que poderia ser feito para facilitar a criação de novos gráficos conforme novos inputs de dados fossem registrados?

```python
import os

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sinasc_2019 = pd.read_csv('./input/SINASC_RO_2019.csv')
sinasc_2019.head(5)
```

```python
sinasc_2019 = sinasc_2019[['IDADEMAE', 'SEXO', 'APGAR1', 
                           'APGAR5', 'PESO', 'CONSULTAS', 
                           'DTNASC', 'GESTACAO', 'GRAVIDEZ', 
                           'ESCMAE', 'IDADEPAI']]
sinasc_2019.head(5)
```

```python
sinasc_2019['DTNASC'] = pd.to_datetime(sinasc_2019['DTNASC'])
sinasc_2019['DTNASC'].head(5)
```

```python
sinasc_2019['MES'] = sinasc_2019['DTNASC'].dt.month
sinasc_2019['MES'] = sinasc_2019['MES'].map({1: 'JAN', 2: 'FEV', 3: 'MAR', 4:'ABR',
                        5: 'MAI', 6: 'JUN', 7: 'JUL', 8: 'AGO',
                        9: 'SET', 10: 'OUT', 11: 'NOV', 12: 'DEZ'})
sinasc_2019['MES'].head(5)
```

```python
dfs_per_month = {month: df_month for month, df_month in sinasc_2019.groupby('MES')}
dfs_per_month['JAN'].head(5)
```

```python
for month in dfs_per_month:
    if os.path.exists('./input/SINASC_RO_2019_{}.csv'.format(month)):
        continue
    else:
        dfs_per_month[month].drop('MES', axis=1).to_csv('./input/SINASC_RO_2019_{}.csv'.format(month))
```

```python
months = ['MAR', 'ABR', 'MAI', 'JUN', 'DEZ']
data = dict()

for month in months:
    path = './input/SINASC_RO_2019_{}.csv'.format(month)
    data.update({month: pd.read_csv(path)})

data['MAR']['DTNASC'].head(5)
```

```python
for month in months:
    data[month]['DTNASC'] = pd.to_datetime(data[month]['DTNASC'])
    data[month]['DIA_SEMANA'] = data[month]['DTNASC'].dt.dayofweek
    data[month]['DIA_SEMANA'] = data[month]['DIA_SEMANA'].map({0: 'SEG', 1: 'TER', 2: 'QUA', 3: 
                                                               'QUI', 4: 'SEX', 5: 'SAB', 6: 'DOM'})

    fig, ax = plt.subplots(figsize=(15, 5))
    agg_data = pd.pivot_table(data[month], 
                              values='IDADEMAE', 
                              index='DIA_SEMANA', 
                              columns='SEXO', 
                              aggfunc='count')
    sns.lineplot(agg_data, ax=ax)
    ax.set_xlabel('Dias da Semana')
    ax.set_ylabel('Quantidade de Nascimentos')
    ax.set_title('Quantidade de Nascimento por Dia da Semana Dividido por Sexo\n({})'.format(month))
    plt.savefig('./output/plot_{}.png'.format(month))
```

O processo poderia ser automizado através de um script que recebe como argumentos os meses desejados, então ele automaticamente ordena os dados por mês, divide-os em subconjuntos de dados e então salva plots dos mesmos.
