# Classificação de Atividade Humana com PCA

Vamos trabalhar com a base da demonstração feita em aula, mas vamos explorar um pouco melhor como é o desempenho da árvore variando o número de componentes principais.

````python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier

from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

from timeit import timeit
````

````python
# Definindo caminho dos dados
filename_features = "../Mod_16/data_tarefa_1/UCI HAR Dataset/features.txt"
filename_labels = "../Mod_16/data_tarefa_1/UCI HAR Dataset/activity_labels.txt"

filename_subtrain = "../Mod_16/data_tarefa_1/UCI HAR Dataset/train/subject_train.txt"
filename_xtrain = "../Mod_16/data_tarefa_1/UCI HAR Dataset/train/X_train.txt"
filename_ytrain = "../Mod_16/data_tarefa_1/UCI HAR Dataset/train/y_train.txt"

filename_subtest = "../Mod_16/data_tarefa_1/UCI HAR Dataset/test/subject_test.txt"
ffilename_xtest = "../Mod_16/data_tarefa_1/UCI HAR Dataset/test/X_test.txt"
filename_ytest = "../Mod_16/data_tarefa_1/UCI HAR Dataset/test/y_test.txt"

# Lendo dados
features = pd.read_csv(filename_features, header=None, names=['nome_var'], sep="#")
labels = pd.read_csv(filename_labels, sep=r'\s+', header=None, names=['cod_label', 'label'])

subject_train = pd.read_csv(filename_subtrain, header=None, names=['subject_id'])
X_train = pd.read_csv(filename_xtrain, sep=r'\s+', header=None, names=features['nome_var'].tolist())
y_train = pd.read_csv(filename_ytrain, header=None, names=['cod_label'])

subject_test = pd.read_csv(filename_subtest, header=None, names=['subject_id'])
X_test = pd.read_csv(ffilename_xtest, sep=r'\s+', header=None, names=features['nome_var'].tolist())
y_test = pd.read_csv(filename_ytest, header=None, names=['cod_label'])

# Repartindo base de dados para criar base de validação
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=.4)
````

## PCA com variáveis padronizadas

Reflexão sobre a escala das variáveis:

**Variáveis em métricas muito diferentes** podem interferir na análise de componentes principais. Lembra que variância é informação pra nós? Pois bem, tipicamente se há uma variável monetária como salário, vai ter uma ordem de variabilidade bem maior que número de filhos, tempo de emprego ou qualquer variável dummy. Assim, as variáveis de maior variância tendem a "dominar" a análise. Nesses casos é comum usar a padronização das variáveis.

Faça duas análises de componentes principais para a base do HAR - com e sem padronização e compare:

- A variância explicada por componente
- A variância explicada acumulada por componente
- A variância percentual por componente
- A variância percentual acumulada por componente
- Quantas componentes você escolheria, em cada caso para explicar 90% da variância?

````python
%%time
# Função para padronizar variáveis
def padroniza(s):
    if s.std() > 0:
        s = (s - s.mean())/s.std()
    return s
````

````python
X_train_pad = pd.DataFrame(X_train).apply(padroniza, axis=0)
X_valid_pad = pd.DataFrame(X_valid).apply(padroniza, axis=0)
X_test_pad = pd.DataFrame(X_test).apply(padroniza, axis=0)
````

````python
# Visualizando tamanho das bases 
print(f'Tamanho base de treino: {X_train.shape}')
print(f'Tamanho base de validação: {X_valid.shape}')
print(f'Tamanho base de teste: {X_test.shape}')
````

````python
# Visualizando tamanho das bases padronizadas
print(f'Tamanho base de treino com padronização: {X_train_pad.shape}')
print(f'Tamanho base de validação com padronização: {X_valid_pad.shape}')
print(f'Tamanho base de teste com padronização: {X_test_pad.shape}')
````

````python
%%time

# Criando dois objetos PCA, cp (com padronização) e sp (sem padronização)
pca_cp = PCA() 
pca_sp = PCA() 

# Treinando modelos
cp_princomp = pca_cp.fit(X_train_pad)  
sp_princomp = pca_sp.fit(X_train)      

print('Para os primeiros 10 componentes, segue a comparação abaixo:')

print('Variância explicada por componente:')
print('Com padronização:', end=' ')
print(cp_princomp.explained_variance_[:10])  
print('Sem padronização:', end=' ')
print(sp_princomp.explained_variance_[:10]) 
print('='*100)

print('Variância explicada acumulada por componente:')
print('Com padronização:', end=' ')
print(cp_princomp.explained_variance_.cumsum()[:10]) 
print('Sem padronização:', end=' ')
print(sp_princomp.explained_variance_.cumsum()[:10])  
print('='*100)

print('Variância explicada percentual por componente:')
print('Com padronização:', end=' ')
print(cp_princomp.explained_variance_ratio_[:10])  
print('Sem padronização:', end=' ')
print(sp_princomp.explained_variance_ratio_[:10])  
print('='*100)

print('Variância explicada percentual acumulada por componente:')
print('Com padronização:', end=' ')
print(cp_princomp.explained_variance_ratio_.cumsum()[:10])  
print('Sem padronização:', end=' ')
print(sp_princomp.explained_variance_ratio_.cumsum()[:10])  
````

````python
# Visualizando graficamente

# Definindo tamanho da figura
plt.figure(figsize=(12, 6))

# Plotando 1º gráfico
plt.subplot(1, 2, 1)
plt.plot(cp_princomp.explained_variance_ratio_[:10], label="Com Padronização", marker='o')
plt.plot(sp_princomp.explained_variance_ratio_[:10], label="Sem Padronização", marker='s')
plt.title('Variância Explicada Percentual por Componente')
plt.xlabel('Componentes Principais')
plt.ylabel('Percentual de Variância Explicada')
plt.legend()

# Plotando 2º gráfico
plt.subplot(1, 2, 2)
plt.plot(cp_princomp.explained_variance_ratio_.cumsum()[:10], label="Com Padronização", marker='o')
plt.plot(sp_princomp.explained_variance_ratio_.cumsum()[:10], label="Sem Padronização", marker='s')
plt.title('Variância Explicada Percentual Acumulada por Componente')
plt.xlabel('Componentes Principais')
plt.ylabel('Percentual Acumulado de Variância Explicada')
plt.legend()

# Ajustes e visualização
plt.tight_layout()
plt.show()
````

````python
# Coletando o percentual de variância explicada acumulada por componente
cp_perc_acum = cp_princomp.explained_variance_ratio_.cumsum()
sp_perc_acum = sp_princomp.explained_variance_ratio_.cumsum()

# Encontrando quantos componentes são necessários para atingir 90% de variância explicada
l1 = list()
for i in range(0, len(cp_perc_acum)):
    if cp_perc_acum[i] <= .902:
        l1.append(i)
        
# Encontrando quantos componentes são necessários para atingir 90% de variância explicada
l2 = list()
for i in range(0, len(sp_perc_acum)):
    if sp_perc_acum[i] <= .902:
        l2.append(i)

# Visualizando resposta
print(f'Para modelagem com padronização seria necessário utilizar {l1[-1]} componentes para explicar {cp_perc_acum[l1[-1]]*100:.2f}% da variância')
print(f'Para modelagem sem padronização seria necessário utilizar {l2[-1]} componentes para explicar {sp_perc_acum[l2[-1]]*100:.2f}% da variância')
````

## Árvore com PCA

Faça duas uma árvore de decisão com 10 componentes principais - uma com base em dados padronizados e outra sem padronizar. Utilize o ```ccp_alpha=0.001```.

Compare a acurácia na base de treino e teste.

````python
# Função para rodar análise de componentes principais
def run_pca(estimator, n_components, *args):
    # Extração dos parâmetros
    pc_train, pc_valid, pc_test, y_train, y_valid, y_test = args
    # Definindo nome das colunas
    cols = ['cp'+str(x+1) for x in list(range(n_components))]

    # Selecionando componentes para treinar e avaliar o modelo
    pc_train_selected = pd.DataFrame(pc_train[:,:n_components], columns=cols)
    pc_valid_selected = pd.DataFrame(pc_valid[:,:n_components], columns=cols)
    pc_test_selected = pd.DataFrame(pc_test[:,:n_components], columns=cols)

    # Treinando o modelo
    trained_model_pca = estimator.fit(pc_train, y_train)

    # Avaliando o modelo nas bases
    train_score_pca = cross_val_score(trained_model_pca, pc_train_selected, y_train, cv=5, scoring='accuracy')
    valid_score_pca = cross_val_score(trained_model_pca, pc_valid_selected, y_valid, cv=5, scoring='accuracy')
    test_score_pca = cross_val_score(trained_model_pca, pc_test_selected, y_test, cv=5, scoring='accuracy')

    return train_score_pca, valid_score_pca, test_score_pca
````

````python
%%time

# Criando objeto de árvore de classificação
clf = DecisionTreeClassifier(random_state=314, ccp_alpha=.001)

# Criando matrizes de autovalores para cada base padronizada
cp_pc_train = cp_princomp.transform(X_train_pad)
cp_pc_valid = cp_princomp.transform(X_valid_pad)
cp_pc_test = cp_princomp.transform(X_test_pad)

# Rodando PCA com as matrizes
cp_train_score_pca, cp_valid_score_pca, cp_test_score_pca = run_pca(clf, 10, cp_pc_train, cp_pc_valid, cp_pc_test, y_train, y_valid, y_test)

# Criando matrizes de autovalores para cada base não padronizada
sp_pc_train = sp_princomp.transform(X_train)
sp_pc_valid = sp_princomp.transform(X_valid)
sp_pc_test = sp_princomp.transform(X_test)

# Rodando PCA com as matrizes
sp_train_score_pca, sp_valid_score_pca, sp_test_score_pca = run_pca(clf, 10, sp_pc_train, sp_pc_valid, sp_pc_test, y_train, y_valid, y_test)

# Visualizando melhores acurácias por base
print(f'Melhor acurácia na base de treino com padronização: {cp_train_score_pca.max()*100:.2f}%')
print(f'Melhor acurácia na base de treino sem padronização: {sp_train_score_pca.max()*100:.2f}%')
print('='*100)
print(f'Melhor acurácia na base de validação com padronização: {cp_valid_score_pca.max()*100:.2f}%')
print(f'Melhor acurácia na base de validação sem padronização: {sp_valid_score_pca.max()*100:.2f}%')
print('='*100)
print(f'Melhor acurácia na base de teste com padronização: {cp_test_score_pca.max()*100:.2f}%')
print(f'Melhor acurácia na base de teste sem padronização: {sp_test_score_pca.max()*100:.2f}%')
````
