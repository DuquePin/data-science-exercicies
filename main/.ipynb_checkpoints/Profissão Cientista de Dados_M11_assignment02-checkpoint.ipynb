### Observação
Assim como no exercícios anterior, vou carregar uma base de dados diferente por questões éticas.

```python
# Base de dados
from sklearn.datasets import fetch_california_housing

# Para manipulação de dados e cálculos
import pandas as pd
import numpy as np
from math import pow

# Para visualização de dados
import seaborn as sns
import matplotlib.pyplot as plt

# Para manipulação e visualização do modelo
from sklearn.tree import DecisionTreeRegressor, export_graphviz
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import train_test_split

import graphviz

# Extraindo dados
housing = fetch_california_housing(as_frame=True)
housing_frame = housing.frame

# Dividindo base de dados em target x features
y = housing_frame['MedHouseVal']
X = housing_frame.copy().drop('MedHouseVal', axis=1)
```

```python
X.head()
```

```python
y.head()
```

### 1. Execute os passos do exercício anterior, até que você tenha uma árvore de regressão predizendo o valor do imóvel na base de treinamento.

Assim como foi feito no exercício anterior, vamos remover a variável **AveOccup** devido a sua baixa relevância para o conjunto de dados.

```python
X.drop('AveOccup', axis=1, inplace=True)
X.head(5)
```

Agora, vamos treinar o melhor modelo encontrado através do gráfico de comparação do **r^2** e **mse** das árvores de regressão criadas no exercício anterior.

```python
def r2(y, y_pred):
    return 1 - (len(y) * mse(y, y_pred) / np.sum([pow(i - np.mean(y), 2) for i in y]))
```

```python
# Dividindo base de dados em base de treino e teste
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=100)

# Criando modelo
regtree = DecisionTreeRegressor(max_depth=9, random_state=100)

# Treinando modelo com conjunto de treino
regtree.fit(x_train, y_train)

# Criando previsões com o modelo treinado
y_pred = regtree.predict(x_test)

# Colhendo r^2 e mean square erro do modelo treinado
print(r2(y_test, y_pred))
print(mse(y_test, y_pred))
```

### 2.  Calcule o caminho indicado pelos CCP-alfas dessa árvore.

```python
path = regtree.cost_complexity_pruning_path(x_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

tab = pd.DataFrame({'ccp_alpha': ccp_alphas, 'total_leaf_impurity': impurities})
tab.head()
```

### 3. Paca cada valor de alpha obtido no item 2, treine uma árvore com o respectivo alfa, e guarde essa árvore em uma lista.

Com o intuito de cruzar a melhor profundidade ao melhor valor de alfa, vamos primeiro comparar os alfas de uma árvore de decisão com profundidade 9, em seguida vamos criar árvores de profundidades variadas e comparar qual delas apresenta o melhor desempenho.

```python
mses, R2 = [], []

for ccp_alpha in tab['ccp_alpha']:
    tmp_regtree = DecisionTreeRegressor(ccp_alpha=ccp_alpha, max_depth=9, random_state=100)
    tmp_regtree.fit(x_train, y_train)
    predict = tmp_regtree.predict(x_test)
    mses.append(mse(y_test, predict))
    R2.append(r2(y_test, predict))

tab['mean_square_error'] = mses
tab['r^2'] = R2
tab.head()
```

```python
fig1, ax1 = plt.subplots(figsize=(12, 8))

ax1.plot(tab['ccp_alpha'], tab['r^2'], 'g-', label='R^2')
ax1.set_xlabel('CCP Alpha')
ax1.set_xticks([value * .01 for value in range(0, 44, 3)])
ax1.set_ylabel('R2', color='g')
ax1.set_yticks([value * .1 for value in range(0, 15, 1)])

ax2 = ax1.twinx()
ax2.plot(tab['ccp_alpha'], tab['mean_square_error'], 'b-', label='MSE')
ax2.set_ylabel('Mean Square Error', color='b')
ax2.set_yticks([value * .1 for value in range(0, 15, 1)])

ax3 = ax1.twiny()
ax3.plot(tab['ccp_alpha'], tab['total_leaf_impurity'], 'r-', label='Impurities')
ax3.set_xlabel('Total Leafs Impurities', color='red')
ax3.set_xticks([value * .01 for value in range(0, 50, 5)])

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
ax3.legend(loc='upper center')

plt.show()
```

Vamos visualizar a relação dessas variáveis de forma mais isoladas.

```python
plt.close()

fig2, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), sharex=True)

sns.lineplot(data=tab, x='ccp_alpha', y='r^2', ax=axes[0, 0])
sns.lineplot(data=tab, x='ccp_alpha', y='mean_square_error', ax=axes[0, 1])
sns.lineplot(data=tab, x='ccp_alpha', y='total_leaf_impurity', ax=axes[1, 0])
sns.lineplot(data=tab, x='ccp_alpha', y='ccp_alpha', ax=axes[1, 1])

for ax in axes.flat:
    ax.set_xticks([value * .01 for value in range(0, 42, 3)]) 
    
plt.tight_layout()
plt.show()
```
