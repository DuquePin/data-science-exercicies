A *Nota Fiscal Paulista*  um programa de incentivo do governo do estado de So Paulo, que retorna um pedacinho do ICMS ao consumidor sempre que solicita o registro do seu CPF na nota. Nesse programa, o contribuinte pode direcionar seus crditos a uma ONG, e quando faz isso, sempre que faz esse consumidor cadastrado faz uma compra, os crditos de notas emitidas pelo mesmo estabelecimento no identificadas (notas sem CPF) so "arrastados" para a ONG na forma de uma doao.

Clique no link, caso queira saber mais sobre a [AMA](https://www.ama.org.br/site/).<br>
Clique no link, caso queira saber como o cidado pode ser um [doador da NFP](https://doacao1.ama.org.br/sitenfp).

Essa base possui dados das notas cujos contribuintes direcionaram seus crditos  AMA. Seus campos esto descritos abaixo:

|Campo|Descrio|
|:-|:-|
|CNPJ emit.| CNPJ do emitente da nota|
|Emitente| Nome fantasia do emitente da nota|

|categoria| Categorizao da nota |

**OBJETIVO:** Algumas notas no do retorno, o que pode ocorrer por diversos motivos, um deles seria a ocorrncia de produtos no incentivados como cigarros e bebidas alclicas. O nosso objetivo  tentar prever que tipo de nota tem maior ou menor propenso a fornecer crditos.

Pode haver uma variabilidade difcil de explicar, em parte por causa de uma freqncia de notas inconstante no tempo (por diversos motivos), em parte por haverem informaes importantes, protegidas pela LGPD, no constantes na base, que pode interferir no crdito gerado, o que vai ocasionar uma variabilidade difcil de se explicar. *Welcome to the jungle*, A vida real  assim :)

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_pickle('./base_nfp.pkl')
df.head()
````

## Anlise da probabilidade de retorno

Considere somente os dados de janeiro de 2020 em diante para esta anlise.

- Analise a proporo de notas que tiveram retorno>0 por categoria de estabelecimento. Esta anlise pode ser por tabela ou grfico.

````python
df_sub = df.loc[(df['Data Emisso'].dt.year >= 2020)].copy()
df_sub.head(5)
````

````python
# Agrupando dados por categoria, obtendo o total de eventos com 'count' e o total de retorno positivo de crdito (1) com 'sum'
tab_cat = df.groupby('categoria', observed=False)['flag_credito'].agg(['count', 'sum'])

# Renomeando as colunas
tab_cat.columns = ['total', 'retorno_positivo']

# Calculando quantidade de eventos de retorno negativo
tab_cat['retorno_negativo'] = tab_cat['total'] - tab_cat['retorno_positivo']

# Calculado o total de eventos e no eventos entre as categorias
total_eventos = tab_cat['retorno_positivo'].sum()
total_no_eventos = tab_cat['retorno_negativo'].sum()

# Calculando distribuio relativa dos eventos e no eventos entre as categorias (contribuio para o total)
tab_cat['dist_rel_retorno_positivo'] = tab_cat['retorno_positivo'] / total_eventos
tab_cat['dist_rel_retorno_negativo'] = tab_cat['retorno_negativo'] / total_no_eventos

# Visualizando
tab_cat
````

````python
# Plotando o grfico com valores absolutos
tab_cat[['retorno_positivo', 'retorno_negativo']].plot(kind='bar', 
                                                       figsize=(12,6), 
                                                       title='Frequncia Absoluta de Retornos Positivos e Negativos por Categoria',
                                                       ylabel='Nmero de Ocorrncias',
                                                       xlabel='Categoria de Estabelecimento'
                                                      )
plt.show()
````

````python
# Plotando o grfico com distribuies relativas (contribuio percentual para o total de cada tipo de evento)
perc = tab_cat[['dist_rel_retorno_positivo', 'dist_rel_retorno_negativo']] * 100
perc.plot(kind='bar', 
          figsize=(12,6), 
          title='Distribuio Relativa de Retornos Positivos e Negativos entre as Categorias',
          ylabel='Porcentagem (%)',
          yticks=range(0, 30, 4),
          xlabel='Categoria de Estabelecimento'
          )
plt.show()
````

````python
# Calculando a proporo de eventos positivos e negativos dentro de cada categoria
tab_cat['prop_retorno_positivo'] = tab_cat['retorno_positivo'] / tab_cat['total']
tab_cat['prop_retorno_negativo'] = tab_cat['retorno_negativo'] / tab_cat['total']

tab_cat
````

````python
# Plotando o grfico com propores dentro de cada categoria
perc = tab_cat[['prop_retorno_positivo', 'prop_retorno_negativo']] * 100
perc.plot(kind='bar', 
          figsize=(12,6), 
          title='Proporo de Retornos Positivos e Negativos dentro de Cada Categoria',
          ylabel='Porcentagem (%)',
          yticks=range(0, 100, 10),
          xlabel='Categoria de Estabelecimento'
          )
plt.show()
````

## Insights

Com base nos grficos acima, vemos que o estabelecimento que possui a maior ocorrncia de retorno positivo de crdito em refercia ao total de notas emitidas pelo mesmo  **Alimentos** com mais de 80% das suas notas concedendo retorno positivo. Do outro lado do espectro, a categoria de **Auto posto** possui mais de 90% das suas notas com retorno negativo, algo que parece intuitivo considerando que entre as categorias, o Auto posto seria a mais comum de vender produtos que no fornecem retorno positivo como cigarros e bebidas alclicas. Observando o grfico da distribuio relativa dos eventos entre as categorias, vemos que a categoria **Mercado** se sobressai sobre as demais compondo mais de 20% das ocorrncias totais de eventos com retorno de crdito positivo, mesmo que em nmero absolutos ela possui a maior quantidade de retorno negativo de crdito, conforme observando no primeiro grfico.

## Clculo do WOE

- Calcule o WOE das categorias, sendo que o evento em questo  a nota ter retorno>0.
- Analise este WOE por tabela e por grfico.

````python
# Calculando razo de chances (rc)
tab_cat['rc'] = tab_cat['dist_rel_retorno_positivo'] / tab_cat['dist_rel_retorno_negativo']

# Calculando weight of evidence (woe)
tab_cat['woe'] = np.log(tab_cat['rc'])

# Visualizando
tab_cat
````

````python
tab_cat['woe'].plot(kind='barh', 
                    figsize=(12,6), 
                    title='WOE por Categoria de Estabelecimento',
                    ylabel='Categoria de Estabelecimento',
                    xlabel='Valor de WOE'
)        
plt.show()
````

## Clculo do *Information Value*

Calcule o *IV* dessa varivel.

````python
iv = sum((tab_cat['dist_rel_retorno_positivo'] - tab_cat['dist_rel_retorno_negativo']) * tab_cat['woe'])
print('O Informattion Value da varivel categoria  {:.2f}'.format(iv))
````
