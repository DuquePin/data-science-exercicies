Carregue a base ```NFP.ftr```. Essa base é proveniente do projeto #AMABiliDados, e contém dados da nota fiscal paulista cadastrados para doação automática para a AMA, Associação de Amigos do Autista. 

A *Nota Fiscal Paulista* é um programa de incentivo do governo do estado de São Paulo, que retorna um pedacinho do ICMS ao consumidor sempre que solicita o registro do seu CPF na nota. Nesse programa, o contribuinte pode direcionar seus créditos a uma ONG, e quando faz isso, sempre que faz esse consumidor cadastrado faz uma compra, os créditos de notas emitidas pelo mesmo estabelecimento não identificadas (notas sem CPF) são "arrastados" para a ONG na forma de uma doação.

Clique no link, caso queira saber mais sobre a [AMA](https://www.ama.org.br/site/).<br>
Clique no link, caso queira saber como o cidadão pode ser um [doador da NFP](https://doacao1.ama.org.br/sitenfp).

Essa base possui dados das notas cujos contribuintes direcionaram seus créditos à AMA. Seus campos estão descritos abaixo:

|Campo|Descrição|
|:-|:-|
|CNPJ emit.| CNPJ do emitente da nota|
|Emitente| Nome fantasia do emitente da nota|
|No.| Número da nota fiscal|
|Data Emissão| Data de emissão da nota fiscal|
|Valor NF| Valor da nota fiscal|
|Data Registro| Data de registro no sistema da NFP|
|Créditos| Valor dos créditos (doação)|
|Situação do Crédito| Se o crédito já foi pago, está sendo processado etc.|
|Ano| Ano da emissão da nota|
|Semestre| Semestre da emissão da nota|
|Retorno| Valor do crédito dividido pelo valor da nota|
|flag_credito| Indicadora se a nota possui crédito positivo|
|categoria| Categorização da nota |

**OBJETIVO:** Algumas notas não dão retorno, o que pode ocorrer por diversos motivos, um deles seria a ocorrência de produtos não incentivados como cigarros e bebidas alcólicas. O nosso objetivo é tentar prever que tipo de nota tem maior ou menor propensão a fornecer créditos.

**Observação**: Esta é uma base real, havendo características de acordo. Pode haver uma variabilidade difícil de explicar, em parte por causa de uma freqüência de notas inconstante no tempo (por diversos motivos), em parte por haverem informações importantes, protegidas pela LGPD, não constantes na base, que pode interferir no crédito gerado, o que vai ocasionar uma variabilidade difícil de se explicar. *Welcome to the jungle*, A vida real é assim :)

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_pickle('./base_nfp.pkl')
df.head()
````

## Análise da probabilidade de retorno

Considere somente os dados de janeiro de 2020 em diante para esta análise.

- Analise a proporção de notas que tiveram retorno>0 por categoria de estabelecimento. Esta análise pode ser por tabela ou gráfico.

````python
df_sub = df.loc[(df['Data Emissão'].dt.year >= 2020)].copy()
df_sub.head(5)
````

````python
# Agrupando dados por categoria, obtendo o total de eventos com 'count' e o total de retorno positivo de crédito (1) com 'sum'
tab_cat = df.groupby('categoria', observed=False)['flag_credito'].agg(['count', 'sum'])

# Renomeando as colunas
tab_cat.columns = ['total', 'retorno_positivo']

# Calculando quantidade de eventos de retorno negativo
tab_cat['retorno_negativo'] = tab_cat['total'] - tab_cat['retorno_positivo']

# Calculado o total de eventos e não eventos entre as categorias
total_eventos = tab_cat['retorno_positivo'].sum()
total_não_eventos = tab_cat['retorno_negativo'].sum()

# Calculando distribuição relativa dos eventos e não eventos entre as categorias (contribuição para o total)
tab_cat['dist_rel_retorno_positivo'] = tab_cat['retorno_positivo'] / total_eventos
tab_cat['dist_rel_retorno_negativo'] = tab_cat['retorno_negativo'] / total_não_eventos

# Visualizando
tab_cat
````

````python
# Plotando o gráfico com valores absolutos
tab_cat[['retorno_positivo', 'retorno_negativo']].plot(kind='bar', 
                                                       figsize=(12,6), 
                                                       title='Frequência Absoluta de Retornos Positivos e Negativos por Categoria',
                                                       ylabel='Número de Ocorrências',
                                                       xlabel='Categoria de Estabelecimento'
                                                      )
plt.show()
````

````python
# Plotando o gráfico com distribuições relativas (contribuição percentual para o total de cada tipo de evento)
perc = tab_cat[['dist_rel_retorno_positivo', 'dist_rel_retorno_negativo']] * 100
perc.plot(kind='bar', 
          figsize=(12,6), 
          title='Distribuição Relativa de Retornos Positivos e Negativos entre as Categorias',
          ylabel='Porcentagem (%)',
          yticks=range(0, 30, 4),
          xlabel='Categoria de Estabelecimento'
          )
plt.show()
````

````python
# Calculando a proporção de eventos positivos e negativos dentro de cada categoria
tab_cat['prop_retorno_positivo'] = tab_cat['retorno_positivo'] / tab_cat['total']
tab_cat['prop_retorno_negativo'] = tab_cat['retorno_negativo'] / tab_cat['total']

tab_cat
````

````python
# Plotando o gráfico com proporções dentro de cada categoria
perc = tab_cat[['prop_retorno_positivo', 'prop_retorno_negativo']] * 100
perc.plot(kind='bar', 
          figsize=(12,6), 
          title='Proporção de Retornos Positivos e Negativos dentro de Cada Categoria',
          ylabel='Porcentagem (%)',
          yticks=range(0, 100, 10),
          xlabel='Categoria de Estabelecimento'
          )
plt.show()
````

## Insights

Com base nos gráficos acima, vemos que o estabelecimento que possui a maior ocorrência de retorno positivo de crédito em referêcia ao total de notas emitidas pelo mesmo é **Alimentos** com mais de 80% das suas notas concedendo retorno positivo. Do outro lado do espectro, a categoria de **Auto posto** possui mais de 90% das suas notas com retorno negativo, algo que parece intuitivo considerando que entre as categorias, o Auto posto seria a mais comum de vender produtos que não fornecem retorno positivo como cigarros e bebidas alcólicas. Observando o gráfico da distribuição relativa dos eventos entre as categorias, vemos que a categoria **Mercado** se sobressai sobre as demais compondo mais de 20% das ocorrências totais de eventos com retorno de crédito positivo, mesmo que em número absolutos ela possui a maior quantidade de retorno negativo de crédito, conforme observando no primeiro gráfico.

## Cálculo do WOE

- Calcule o WOE das categorias, sendo que o evento em questão é a nota ter retorno>0.
- Analise este WOE por tabela e por gráfico.

````python
# Calculando razão de chances (rc)
tab_cat['rc'] = tab_cat['dist_rel_retorno_positivo'] / tab_cat['dist_rel_retorno_negativo']

# Calculando weight of evidence (woe)
tab_cat['woe'] = np.log(tab_cat['rc'])

# Visualizando
tab_cat
````

````python
tab_cat['woe'].plot(kind='barh', 
                    figsize=(12,6), 
                    title='WOE por Categoria de Estabelecimento',
                    ylabel='Categoria de Estabelecimento',
                    xlabel='Valor de WOE'
)        
plt.show()
````

## Cálculo do *Information Value*

Calcule o *IV* dessa variável.

````python
iv = sum((tab_cat['dist_rel_retorno_positivo'] - tab_cat['dist_rel_retorno_negativo']) * tab_cat['woe'])
print('O Informattion Value da variável categoria é {:.2f}'.format(iv))
````
