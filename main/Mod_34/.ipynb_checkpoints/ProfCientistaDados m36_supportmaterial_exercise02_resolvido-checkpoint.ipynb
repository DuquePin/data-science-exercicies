## Criando variveis

**Variveis temporais**:  possvel que haja alguma regra de liberao de crditos de acordo com o dia do ms, o dia da semana, fim de semana (sim ou no), ou at alteraes temporais da regra. Vamos criar as seguintes variveis para testar estas hipteses:

- Dia da semana
- Dummy para fim de semana
- Dia do ms
- Ms 
- Trimestre
- Ano (j na base)

**Valor da nota**: Vamos criar tambm algumas variveis para resumir o valor da nota, e posteriormente criar uma intuio sobre categorizaes de variveis contnuas.

- Crie categorizaes da varivel ```Valor NF```, por quants, em 5, 10, 20 e 50 categorias.

````python
# Criando varivel dia da semana
df['Dia da Semana'] = df['Data Registro'].dt.weekday.map({i: dia for i, dia in enumerate(['Segunda-feira', 'Tera-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sbado', 'Domingo'])})

# Criando flag de fim de semana
df['flag_fds'] = 0
df.loc[df['Dia da Semana'].isin(['Sbado', 'Domingo']), 'flag_fds'] = 1

# Criando varivel do dia do ms
df['Dia'] = df['Data Registro'].dt.day

# Criando varivel do ms
df['Ms'] = df['Data Registro'].dt.month.map({i: mes for i, mes in enumerate(['Janeiro', 'Fevereiro', 'Maro', 'Abril', 'Maio', 'Agosto', 'Junho', 'Julho', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'], start=1)})

# Criando varivel do trimestre
q = [3, 6, 9, 12]
df['Trimestre'] = 0
df.loc[df['Data Registro'].dt.month <= q[0] , 'Trimestre'] = 1
df.loc[(df['Data Registro'].dt.month > q[0]) & (df['Data Registro'].dt.month <= q[1]), 'Trimestre'] = 2
df.loc[(df['Data Registro'].dt.month > q[1]) & (df['Data Registro'].dt.month <= q[2]), 'Trimestre'] = 3
df.loc[df['Data Registro'].dt.month > q[2], 'Trimestre'] = 4

df.head(5)
````

````python
# Verificando consistncia da varivel trimestre criada
pd.crosstab(df['Data Registro'].dt.month, df['Trimestre'])
````

````python
# Criando variveis categorizadas da varivel Valor NF, por quants, em 5, 10, 20 e 50 categorias.
df['cat5_Valor NF'] = pd.qcut(df['Valor NF'], 5)
df['cat10_Valor NF'] = pd.qcut(df['Valor NF'], 10)
df['cat20_Valor NF'] = pd.qcut(df['Valor NF'], 20)
df['cat50_Valor NF'] = pd.qcut(df['Valor NF'], 50)

df[['Valor NF', 'cat5_Valor NF', 'cat10_Valor NF', 'cat20_Valor NF', 'cat50_Valor NF']].head(10)
````

## Calcule o IV de todas essas variveis

Sugesto: Construa um metadados e tente utilizar uma funo como a que foi utilizada durante a aula. Voc vai conseguir fazer esse clculo para as variveis uma a uma, mas em casos reais em que vamos mais alm na criao de variveis, podemos ter centenas, ou at milhares de variveis, esses clculos *precisam* ser automticos nesses casos.

````python
def calcular_iv_variavel(df: pd.DataFrame, var_estudada: str, var_evento: str) -> float:
    """
    Calcula o Information Value (IV) de uma varivel em relao a um evento binrio.

    Parmetros:
    - df: DataFrame contendo os dados
    - var_estudada: Nome da varivel categrica a ser analisada
    - var_evento: Nome da varivel binria indicando o evento (0 = 'no evento', 1 = 'evento')

    Retorna:
    - O valor do Information Value (IV)
    """

    tab = df.groupby(var_estudada, observed=False)[var_evento].agg(['count', 'sum'])
    
    tab.columns = ['total', 'retorno_positivo']
    tab['retorno_negativo'] = tab['total'] - tab['retorno_positivo']

    total_eventos = tab['retorno_positivo'].sum()
    total_no_eventos = tab['retorno_negativo'].sum()
    
    tab['dist_retorno_positivo'] = tab['retorno_positivo'] / total_eventos
    tab['dist_retorno_negativo'] = tab['retorno_negativo'] / total_no_eventos

    tab['woe'] = np.log(tab['dist_retorno_positivo'] / tab['dist_retorno_negativo'])

    tab['iv'] = (tab['dist_retorno_positivo'] - tab['dist_retorno_negativo']) * tab['woe']
    
    return tab['iv'].sum()
````

````python
for var in ['Dia da Semana', 'flag_fds', 'Dia', 'Ms', 'Trimestre', 'cat5_Valor NF', 'cat10_Valor NF', 'cat20_Valor NF', 'cat50_Valor NF']:
    iv = calcular_iv_variavel(df, var, 'flag_credito')
    print(f'O Information Value (IV) da varivel {var}  igual a aproximadamente {iv:.2f}')
````

## Conclua

- Quais variveis parecem apresentar maior poder de discriminao segundo o IV?
- Quando categorizamos variveis contnuas parece haver alguma relao entre o nmero de categorias e o IV?

De acordo com a classificao de *Naeem Siddiqi* as variveis com maior poder discriminatrio so:
- cat5_Valor NF
- cat10_Valor NF
- cat50_Valor NF

As demais ou so inteis ou fracas demais.

Conforme observado pela relao acima, parece existir uma tendncia de quanto maior o nmero de categorias numa varivel contnua, maior o seu IV.

## Avalie o WOE e o IV no tempo

Avalie o WOE e o IV no tempo da varivel ```categoria```.

````python
# Contagem de total de observaes e eventos por categoria
grouped = df.groupby('categoria')['flag_credito'].agg(['count', 'sum'])
grouped.columns = ['total', 'positivo']  

# Calcular quantidade de no eventos
grouped['negativo'] = grouped['total'] - grouped['positivo']

# Nmero total de eventos e no eventos
total_eventos = grouped['positivo'].sum()
total_no_eventos = grouped['negativo'].sum()

# Distribuies de evento e no eventos
grouped['evento_dist'] = grouped['positivo'] / total_eventos
grouped['no_evento_dist'] = grouped['negativo'] / total_no_eventos

grouped
````

````python
# Lista vazia para armazenar IV
iv_results = []

# Calculando IV para cada trimestre
for periodo, subset in df.groupby('Trimestre'):
    iv_value = calcular_iv_variavel(df=subset, var_estudada='categoria', var_evento='flag_credito')
    iv_results.append({'Trimestre': periodo, 'IV': iv_value})

# Transformando lista com dict em dataframe
iv_df = pd.DataFrame(iv_results)

# Plotando grfico
plt.figure(figsize=(10, 5))
sns.lineplot(data=iv_df, x='Trimestre', y='IV', marker='o', linestyle='-')
plt.title('Evoluo do Information Value de Categoria por Perodo')
plt.xlabel('Trimestre')
plt.ylabel('Information Value (IV)')
plt.grid()
plt.show()
````

Dicas:<br>

````python
# Trabalhando com recorte temporal a partir de 2020
df_sub = df.loc[(df['Data Emisso'].dt.year >= 2020)].copy()

# Agrupando por data e categoria, somando os eventos positivos 
df_grouped = df_sub[df_sub['flag_credito'] == 1].groupby(['Data Registro', 'categoria']).size().unstack(fill_value=0)

# Calculando a proporo de cada categoria dentro de cada data
df_prop = df_grouped.div(df_grouped.sum(axis=1), axis=0)

# Criando os subplots (um para cada categoria)
num_categorias = len(df_prop.columns)
fig, axes = plt.subplots(nrows=num_categorias, ncols=1, figsize=(12, 2 * num_categorias), sharex=True)

# Criando um grfico separado para cada categoria
for ax, categoria in zip(axes, df_prop.columns):
    ax.plot(df_prop.index, df_prop[categoria], label=categoria, linewidth=2)
    ax.fill_between(df_prop.index, df_prop[categoria], alpha=0.3)
    ax.set_ylabel('Proporo')
    ax.set_title(f'Categoria: {categoria}')
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.5)

# Ajustando layout e rtulos finais
plt.xlabel('Data de Registro')
plt.suptitle('Proporo de Retornos Positivos por Categoria ao Longo do Tempo', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.xticks(rotation=45)

# Exibir o grfico
plt.show()
````
