```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(dict((col, [np.random.randint(1, 5) for _ in range(10)]) for col in ['col_{}'.format(n+1) for n in range(10)]))
df.head(5)
```

```python
df['sqrt_col_1'] = np.sqrt(df['col_1'])
df['sqrt_col_1']
```

```python
df.assign(sqrt_col_2 = lambda x: np.sqrt(x.col_2))
```

```python
def sum_if_odd(x, y):
    if x % 2 != 0:
        return x + y
    else:
        return 0
```

```python
df['sum_col_3_&_4'] = list(map(lambda x, y: sum_if_odd(x, y), df['col_3'], df['col_4']))
df[['col_3', 'col_4', 'sum_col_3_&_4']]
```

```python
df.pipe(np.power, 2)
```
