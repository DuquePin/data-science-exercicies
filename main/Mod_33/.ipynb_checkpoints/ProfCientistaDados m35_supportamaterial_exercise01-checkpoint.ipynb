Com base na tarefa realizada no mdulo anterior, vamos fazer um diagnstico do modelo.

Carregue a base ```previsao_de_renda2.csv```. Separe em uma base de treino e uma base de teste.

Vamos resgatar a melhor verso do modelo e previso de renda que voc fez para esta base. 

- Substitua missings pela mdia
- Rode novamente o modelo na base de treino (ou desenvolva-o caso no tenha o registro guardado). 

Este modelo deve prever a varivel ```renda``` com base nas demais variveis exceto ```data_ref``` e ```index```. J vimos que a varivel renda  melhor modelada com a transformao ```log()```, no se esquea disso.

````python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf

df = pd.read_csv('../Mod_32/previsao_de_renda_II.csv')
df.head()
````

````python
# Substituindo missings pela mdia
df['tempo_emprego'] = df['tempo_emprego'].fillna(df['tempo_emprego'].mean())
df.info()
````

````python
df['data_ref'] = pd.to_datetime(df['data_ref'])
df['data_ref']
````

````python
df['mes'] = df['data_ref'].dt.month
df['mes']
````

````python
# Dividindo base de dados em bases de treino e teste
X = df.drop(['renda'], axis=1)
y = np.log(df['renda'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=412)

df_train = pd.concat([X_train, y_train], axis=1)
df_test = pd.concat([X_test, y_test], axis=1)

df_train.rename({'renda': 'log_renda'}, axis=1, inplace=True)
df_test.rename({'renda': 'log_renda'}, axis=1, inplace=True)

df_train.head(5)
````

````python
reg = smf.ols('log_renda ~ sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_residencia + idade + tempo_emprego + qt_pessoas_residencia', 
              data=df_train).fit()

reg.rsquared_adj
````

## Verifique as suposies do modelo

Faa uma anlise para avaliar as suposies bsicas:

- Os resduos desse modelo possuem distribuio Normal (ou algo prximo)?
- Verifique a independncia dos resduos
    - Faa um grfico dos resduos versus os valores preditos
    - Avalie se h padres dos resduos versus cada uma das variveis do modelo
    - Avalie se o valor mdio dos resduos aparenta ter relao com o ms de referncia
- Avalie se a varincia da varivel resposta parece ser conforme os valores previstos.

Voc considera que as suposies esto atendidas? H algum impacto em eventuais desvios?

````python
df_train['resid'] = reg.resid
df_train['fitted_values'] = reg.fittedvalues
df_train.head(5)
````

````python
# 1. Verificao da normalidade dos resudos
plt.figure(figsize=(12,6))
sns.histplot(df_train['resid'], kde=True, bins=20)
plt.title('Distribuio dos Resduos')
plt.show()
````

Conforme observado pelo grfico acima, os resduos do modelo possuem uma distribuio normal.

````python
# 2. Independncia dos resduos
plt.figure(figsize=(12,6))
sns.scatterplot(x=df_train['fitted_values'], y=df_train['resid'])
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Resduos vs Valores Preditos')
plt.show()
````

````python
df_train['sexo'] = df_train['sexo'].map({'F': 1, 'M': 0})
df_train['posse_de_veiculo'] = df_train['posse_de_veiculo'].map({'N': 0, 'S': 1})
df_train['posse_de_imovel'] = df_train['posse_de_imovel'].map({'N': 0, 'S': 1})
df_train['tipo_residencia'] = df_train['tipo_residencia'].map({'Casa': 1, 'Governamental': 2, 'Aluguel': 3, 'Estdio': 4, 'Com os pais': 5, 'Comunitrio': 6})
````

````python
df_train[['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'qtd_filhos', 'tipo_residencia', 'idade', 'tempo_emprego', 'qt_pessoas_residencia']]
````

````python
# 3. Padres dos resduos vs variveis independentes

# Combinao linear das variveis explicativas do modelo em um array unidimensional
pca = (PCA(n_components=1)
       .fit_transform(df_train[['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'qtd_filhos', 'tipo_residencia', 'idade', 'tempo_emprego', 'qt_pessoas_residencia']])
       .flatten())

# Visualizando grfico
plt.figure(figsize=(12,6))
sns.scatterplot(x=pca, y=df_train['resid'])
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Resduos vs Variveis Explicativas')
plt.show()
````

````python
# Funo para plotar grfico
def plotar_grafico_variavel_explicativa_vs_residuo(var):
    plt.figure(figsize=(12,6))
    sns.scatterplot(x=df_train[var], y=df_train['resid'])
    plt.axhline(y=0, color='r', linestyle='--')
    plt.title('Resduos vs Varivel {}'.format(var))
    plt.show()
````

````python
for var in ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'qtd_filhos', 'tipo_residencia', 'idade', 'tempo_emprego', 'qt_pessoas_residencia']:
    plotar_grafico_variavel_explicativa_vs_residuo(var)
````

Observando pelos grficos acima, as variveis explicativas no parecem demonstrar nenhum padro em sua distribuio afetado pelos resduos, portanto podemos considerar os resduos **independentes**.

````python
# Avaliando se valor mdio dos resduos  afetado pelo ms de referencia
plotar_grafico_variavel_explicativa_vs_residuo('mes')
````

````python
gp = df_train.groupby(by='mes')['resid'].mean()
gp.plot.line()
plt.show()
````

Observando pelos grficos acima, de fato os resduos do modelo so afetados pela varivel explicativa **mes**, contudo, como exclumos essa varivel do modelo, esse permanece inafetado por essa propriedade.

````python
# 4. Homocedasticidade (varincia da varivel resposta com valores preditos)
plt.figure(figsize=(12,6))
sns.scatterplot(x=df_train['fitted_values'], y=np.abs(df_train['resid']))
plt.title('Resduos Absolutos vs Valores Preditos')
plt.show()
````

Observando o grfico acima, no parece existir padro nenhum que indique uma relao entre a varincia da varivel resposta e o resduo do modelo, portanto a suposio de **homecedasticidade** do modelo  verdadeira.

## Outliers

Avalie os *studentized residuals*, verifique se h pontos que parecem ser discrepantes.

Avalie se h pontos influentes.

````python
influence = reg.get_influence()

studentized_residuals = influence.resid_studentized_external

outliers = np.where(np.abs(studentized_residuals) > 2)[0]

distance_cook = influence.cooks_distance[0]

influential_points = np.where(distance_cook > 1)[0]

print('Studentized Residuals (Outliers):', outliers)
print('Pontos influentes (Distncia de Cook > 1):', influential_points)

plt.figure(figsize=(10,5))
plt.scatter(range(len(studentized_residuals)), studentized_residuals, label='Resduos Studentizados')
plt.axhline(y=2, color='r', linestyle='--', label='Limite Superior')
plt.axhline(y=-2, color='r', linestyle='--', label='Limite Inferior')
plt.xlabel('ndice da Observao')
plt.ylabel('Resduo Studentizado')
plt.legend()
plt.title('Resduos Studentizados e Outliers')
plt.show()
````
